<h2 style="color:blue" align="center">Machine Learning With Python</h2>

<h3 style="color:purple" align="center">Activation Functions</h3>

```python
# import libraries/dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# function to plot activation functions

def plot_func(x,y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('Activation Function f(x)')
    plt.grid(True)
    plt.show()
```

```python
x = np.linspace(-10, 10, 100)
```

<h3 style="color:purple" align="center" >Binary (threshold)</h3>

```python
# binary
def binary(x):
    return 1 if x>=0 else 0
```

```python
bn_af = np.vectorize(binary)(np.linspace(-10, 10, 100))
```

```python
plot_func(x,bn_af, 'Binary')
```

<h3 style="color:purple" align="center">Linear</h3>

```python
# linear 
def linear(x):
    return x
```

```python
ln_af = linear(np.linspace(-10, 10, 100))
```

```python
plot_func(x,ln_af, 'Linear')
```

<h3 style="color:purple" align="center">Non-Linear</h3>

<h4 style="color:blue" >Sigmoid</h4>

```python
# sigmoid 
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
sg_af = sigmoid(np.linspace(-10, 10, 100))
```

```python
plot_func(x,sg_af, 'Sigmoid')
```

<h4 style="color:blue" >tanh</h4>

```python
# tanh
def tanh(x):
    return np.tanh(x)
```

```python
tanh_af = tanh(np.linspace(-10, 10, 100))
```

```python
plot_func(x,tanh_af, 'Hyperbolic Tangent')
```

<h4 style="color:blue" >Rectified Linear Unit (ReLU)</h4>

```python
# ReLU
def ReLU(x):
    return np.maximum(0,x)
```

```python
ReLU_af = ReLU(np.linspace(-10, 10, 100))
```

```python
plot_func(x,ReLU_af, 'ReLU')
```

<h4 style="color:blue" >Leaky Rectified Linear Unit (Leaky ReLU)</h4>

```python
# Leaky ReLU
def leaky_ReLU(x):
    return np.maximum(0.1*x,x)
```

```python
LReLU_af = leaky_ReLU(np.linspace(-10, 10, 100))
```

```python
plot_func(x,LReLU_af, 'Leaky ReLU')
```

<h4 style="color:blue" >Parametric Linear Unit (PReLU)</h4>

```python
# Parametric ReLU
def para_ReLU(x, alpha):
    return np.maximum(alpha*x,x)
```

```python
plt.plot(x,para_ReLU(np.linspace(-10, 10, 100), 0.05), label='alpha = 0.05')
plt.plot(x,para_ReLU(np.linspace(-10, 10, 100), 0.01), label='alpha = 0.01')
plt.plot(x,para_ReLU(np.linspace(-10, 10, 100), 0.1), label='alpha = 0.1')
plt.plot(x,para_ReLU(np.linspace(-10, 10, 100), 0.2), label='alpha = 0.2')
plt.title('Parametric ReLU')
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('Parametric ReLU, PReLU(x)')
```

<h4 style="color:blue" >Exponential Linear Unit (ELU)</h4>

```python
#  ELU
def ELU(x, alpha):
    return alpha*(np.exp(x)-1) if x<0 else x
```

```python
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 0.05), label='alpha = 0.05')
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 0.01), label='alpha = 0.01')
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 0.1), label='alpha = 0.1')
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 0.2), label='alpha = 0.2')
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 0.5), label='alpha = 0.5')
plt.plot(x,np.vectorize(ELU)(np.linspace(-10, 10, 100), 1), label='alpha = 1')
plt.title('ELU')
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('Exponential Linear Unit, ELU(x)')
```

<h4 style="color:blue" >Softmax</h4>

```python
#  softmax
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
sftmx = softmax(np.linspace(-10, 10, 100))
```

```python
plot_func(x,sftmx,'Softmax')
```

<h4 style="color:blue" >Swish</h4>

```python
#  softmax
def swish(x):
    return x /(1+np.exp(-x))
```

```python
swsh = swish(np.linspace(-10, 10, 100))
```

```python
plot_func(x,swsh,'Swish')
```

```python
# swish = x*sigmoid
plot_func(x,x*sg_af,'Swish')
```

<h4 style="color:blue" >Gaussian Error Linear Unit (GELU)</h4>

```python
#  Gaussian Error Linear Unit
def GELU(x):
    return 0.5 * x * (1 + np.tanh((np.sqrt(2 / math.pi)) * (x + 0.044715 * x**3)))
```

```python
gelu = GELU(np.linspace(-10, 10, 100))
```

```python
plot_func(x,gelu,'GELU')
```

<h4 style="color:blue" >Scaled Exponential Linear Unit (SELU)</h4>

```python
#  Scaled Exponential Linear Unit
def SELU(x,alpha,lmbd):
    return lmbd*alpha*(np.exp(x)-1) if x<0 else lmbd*x
```

```python
selu = np.vectorize(SELU)(np.linspace(-10, 10, 100),1.6733,1.0507)
```

```python
plot_func(x,selu,'SELU')
```
